```python
#Checking missing values
missing_values = data.isnull().sum().to_frame().reset_index()
```

```python
#figsize
fig, ax = plt.subplots(figsize =(8, 4))

#plotting missing value
plt.bar(missing_values['index'],missing_values[0])
plt.show()
```

# Outliers

## Mean, Median, and Mode are the Central Tendencies used widely to understand data distribution while identifying outliers.

```python
# mean = sum of all elements in set/number os elements in set
# to calculate median the numbers must the in crescent order. When the size of the set is even: 
# median = mean of the central numbers
# if the number of elements in set is odd
# median = central number
```

```python
# Box plot uses the foundation of IQR to calculate outliers from the data.

data['Age']. describe()
```

```python
plt.figure(figsize=(15,7))
sns.boxplot(x = 'Age', data = data, showmeans=True)
plt.title("Boxplot for Age")
plt.show()
```

```python
correlation = data.corr().round(2) # creating a 2-D Matrix with correlation plots
correlation
plt.figure(figsize = (15,8)) #figsize=(width,height)
sns.heatmap(correlation, annot=True, cmap='vlag_r', vmin=-1, vmax=1); #color palette
```
